```python
import nltk
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.tokenize import SpaceTokenizer, TabTokenizer, LineTokenizer
```

1. word_tokenize

```python
sent = input('Sentence: ')
word_tokenize(sent)
```

2. sent_tokenize

```python
sent = input('Sentence: ')
sent_tokenize(sent)
```

3. SpaceTokenizer

```python
sent = input('Sentence: ')
st = SpaceTokenizer()
x = st.tokenize(sent)
print(x)
```

4. TabTokenizer

```python
sent = 'The Chera dynasty \twas \tthe first prominent kingdom \tbased in Kerala.'
print(sent)

tt = TabTokenizer()
x = tt.tokenize(sent)
print(x)
```

5. LineTokenizer

```python
ns = 'Kerala has \nthe lowest positive population growth rate \nin India.'
print(ns)
lt = LineTokenizer()
y = lt.tokenize(ns)
print(y)
```

<b>Write a program to perform stemming, lemmatization and pos tagging on the
following text.<b>

1. eclipses
2. inspiring
3. viewed
4. astronomical
5. unlikely
6. closest
7. approximately
8. during
9. months

```python
import nltk
```

```python
words = ["eclipses", "inspiring", "viewed", "astronomical", "unlikely", "closest", "approximately", "during", "months"]
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
for word in words:
    print(ps.stem(word))
```

```python
from nltk.stem import LancasterStemmer
ls = LancasterStemmer()
for word in words:
    print(ls.stem(word))
```

```python
from nltk.stem import WordNetLemmatizer
wnl = WordNetLemmatizer()
lemmatized_words = [wnl. lemmatize(word) for word in words]
print(lemmatized_words)
```

```python
from nltk import pos_tag
pos_tag(words)
```
